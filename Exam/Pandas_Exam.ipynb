```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Sample DataFrame
data = {
    'Region': ['North', 'South', 'East', 'West', 'North', 'South', 'East', 'West', 'North', 'South'],
    'Sales': [20000, 25000, 30000, 22000, 28000, 27000, 31000, 24000, 26000, 29000],
    'Profit': [4000, 5000, 5500, 4200, 4700, 5300, 6000, 4800, 4500, 5200]
}

df = pd.DataFrame(data)

# User selects two columns for visualization
print("\nAvailable Columns:", list(df.columns))
col1 = input("Enter the first column for visualization: ")
col2 = input("Enter the second column for visualization: ")

if col1 not in df.columns or col2 not in df.columns:
    print("Invalid column selection. Please try again.")
else:
    while True:
        print("\n == Data Visualization ==")
        print("   1. Bar plot")
        print("   2. Line plot")
        print("   3. Scatter plot")
        print("   4. Pie chart")
        print("   5. Box Plot")
        print("   6. Histogram")
        print("   7. Violin Plot")
        print("   8. Stack Plot")
        print("   9. Step Chart")
        print("   10. Exit\n")

        choice = input("Enter your choice: ")

        plt.figure(figsize=(8, 5))

        if choice == '1':
            df.groupby(col1)[col2].sum().plot(kind='bar', color='blue')
            plt.title(f"Bar Plot of {col2} by {col1}")

        elif choice == '2':
            df.plot(x=col1, y=col2, kind='line', marker='o', color='green')
            plt.title(f"Line Plot of {col2} vs {col1}")

        elif choice == '3':
            plt.scatter(df[col1], df[col2], color='red')
            plt.xlabel(col1)
            plt.ylabel(col2)
            plt.title(f"Scatter Plot of {col2} vs {col1}")

        elif choice == '4':
            df.groupby(col1)[col2].sum().plot(kind='pie', autopct='%1.1f%%', startangle=90, cmap='viridis')
            plt.ylabel('')
            plt.title(f"Pie Chart of {col2} by {col1}")

        elif choice == '5':
            sns.boxplot(x=df[col1], y=df[col2])
            plt.title(f"Box Plot of {col2} by {col1}")

        elif choice == '6':
            plt.hist(df[col2], bins=5, color='orange', edgecolor='black')
            plt.xlabel(col2)
            plt.ylabel("Frequency")
            plt.title(f"Histogram of {col2}")

        elif choice == '7':
            sns.violinplot(x=df[col1], y=df[col2])
            plt.title(f"Violin Plot of {col2} by {col1}")

        elif choice == '8':
            plt.stackplot(df.index, df[col2], colors=['purple'], alpha=0.6)
            plt.title(f"Stack Plot of {col2}")

        elif choice == '9':
            plt.step(df[col1], df[col2], color='brown', where='mid')
            plt.xlabel(col1)
            plt.ylabel(col2)
            plt.title(f"Step Chart of {col2} vs {col1}")

        elif choice == '10':
            print("Exiting Visualization...")
            break

        else:
            print("Invalid choice. Please enter a number between 1-10.")
            continue

        plt.xticks(rotation=45)
        plt.grid(axis='y', linestyle='--', alpha=0.7)
        plt.show()
```
